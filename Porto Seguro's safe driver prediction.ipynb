```python
% matplotlib inline
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

### Importing data

```python
#Importing train file
df= pd.read_csv("train.csv")
pd.set_option('display.max_row', 1000)
pd.set_option('display.max_columns',100)
```

### Finding Missing Values

```python
#finding missing values
# For this data set, the missing values are indicated by -1. Hence we shall replace all the -1 by NaN,for python to recognize.
df1=df.replace(-1, np.NaN)
df1.isnull().sum()
df1.head()
```

### Data Cleaning

```python
#Remove the rows with missing values in the columns found.
df1.dropna(subset=["ps_ind_02_cat","ps_ind_04_cat","ps_ind_05_cat","ps_reg_03","ps_car_01_cat","ps_car_02_cat","ps_car_03_cat","ps_car_05_cat","ps_car_07_cat","ps_car_09_cat","ps_car_11","ps_car_12","ps_car_14",], axis=0, inplace= True)
df1.isnull().sum()
df1.shape
```

### Feature Scaling and Feature Selection

```python
# Feature scaling
df1["ps_car_06_cat"]= (df1["ps_car_06_cat"]-df1["ps_car_06_cat"].mean())/df1["ps_car_06_cat"].std()
df1["ps_car_01_cat"]= (df1["ps_car_01_cat"]-df1["ps_car_01_cat"].mean())/df1["ps_car_01_cat"].std()
```

```python
# Converting Categorical variables to Numeric
df1_encoded=pd.get_dummies(data= df1, columns= ["ps_ind_02_cat","ps_ind_04_cat","ps_ind_05_cat","ps_car_01_cat","ps_car_02_cat","ps_car_03_cat","ps_car_04_cat","ps_car_05_cat","ps_car_06_cat","ps_car_07_cat","ps_car_08_cat","ps_car_09_cat","ps_car_10_cat","ps_car_11_cat"])
```

### Dealing with imbalanced data

```python
#Find if the target column is balanced 
df1_encoded["target"].value_counts(dropna= False )
```

```python
#It's highly imbalanced. So, lets balance it
from sklearn.utils import resample
df1_majority = df1_encoded[df1_encoded.target==0]         #Create a dataframe with majority class
df1_minority = df1_encoded[df1_encoded.target==1] 
#Create a dataframe with minority class
df1_minority.shape
df1_majority.shape

#upsample minority class
df1_minority_upsampled = resample(df1_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=90000,    # number of samples to generate
                                 random_state=123) # reproducible results
df1_minority_upsampled.shape
```

```python
# Combine majority class with upsampled minority class
df1_upsampled = pd.concat([df1_majority, df1_minority_upsampled])
```

```python
#check if they are balanced now
df1_upsampled.target.value_counts()
df1_upsampled.head()
```

### Separating Input variables and Target

```python
#separating data and target features from the dataframe
y= df1_upsampled.target
headers=["target"]
y.columns=headers
```

```python
X =df1_upsampled.drop('target', 1)
```

### Dimensionality reduction using PCA

```python
from sklearn.decomposition import PCA as sklearnPCA
sklearn_pca = sklearnPCA(n_components=20)
X = sklearn_pca.fit_transform(X)
```

### Splitting the data 

```python
#Split into training and test sets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
#Applying Randomforest
from xgboost import XGBClassifier
model = XGBClassifier()
```

### Training the model

```python
# Train the model using the training sets and check score
model.fit(X_train,y_train)
```

```python
model.score(X_train, y_train)
```
